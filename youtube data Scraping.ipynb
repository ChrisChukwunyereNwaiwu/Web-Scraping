Import Libraries

```python
import requests
import random
import time
```

```python

# My API key
API_KEY = 'AIzaSyAN0mnmtNQBHece03UfZg3SqdLVSjtzZ1g'

# Function to fetch live chat comments for a video
def fetch_live_chat_comments(video_id):
    chat_comments = []
    api_url = f'https://www.googleapis.com/youtube/v3/commentThreads?key={API_KEY}&textFormat=plainText&part=snippet&videoId={video_id}&maxResults=100'

    response = requests.get(api_url)
    if response.status_code == 200:
        data = response.json()

        for item in data.get('items', []):
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            chat_comments.append(comment)

        # Introduce a delay to avoid overloading the API
        time.sleep(random.uniform(2, 5))

    return chat_comments

# List of YouTube video IDs to retrieve live chat from
video_ids = [
    "UCpUBh7qelVwXS_m2ZH46kjg",  # J Hus - British Rapper Channel
    "UCEzDdNqNkT-7rSfSGSr1hWg",  # BurnaBoy Afrobeats musician's channel
    "n98Ixr0vdcU",  # Odegaard spot on as 10-player Arsenal see off Palace( Video)
    "Uf0LEqdhSW4",  # Wagner boss Prigozhin listed as passenger on crashed plane in Russia - BBC News
    "veMFCFyOwFI",  # The Middle East's cold war, explained - VOX
    "BB3qNWRaxGE",  # Why US gun laws get looser after mass shootings
    "1oqyrflOQFc",  # The Myth of the Gender Wage Gap
    "eqBfSPgsrKw",  # Shame on You for Killing Babies: Republican Lawmaker Loses Cool at Abortion Rights Protesters -Thequint
    "rkzZJAvjIWE",  # Cold Mountain: Union soldiers try to rob and rape Sara - Binge Society
    "GNiMNUcs530"   # Russian armored column massacred | Military Mind - TVP World
]

# Initialize an empty list to store chat comments
all_chat_comments = []

# Fetch comments from each video
for video_id in video_ids:
    comments = fetch_live_chat_comments(video_id)
    all_chat_comments.extend(comments)

# Shuffle the comments
random.shuffle(all_chat_comments)

# Select 500 comments from the shuffled list
selected_comments = all_chat_comments[:500]

# Write the selected comments to a text file
with open("live_chat_dataset.txt", "w", encoding="utf-8") as file:
    for comment in selected_comments:
        file.write(comment + "\n")
```
